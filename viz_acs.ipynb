### ACS visualizations
    

show the relationship between connectivity and various socio-economic stats, like:
- race
- unemployment rate
- median income
- hardship index
- CCVI

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
chicago_df = pd.read_csv("data/chicago_internet.csv",index_col=0,
                      parse_dates=[0])
```

```python
chicago_df
```

```python
plt.figure(figsize=(15, 10), dpi=200)

plt.scatter(x=chicago_df["hh_no_internet(%)"],
            y=chicago_df["total_black(%)"],
           c="red", alpha=0.5)


plt.xlabel("Percentage of Households with No Internet(%)")
plt.ylabel("Percentage of Black Population in each Chicago Community Area (%)")
plt.title("Internet Access by Race (2015-19)")

```

```python
plt.figure(figsize=(15, 10), dpi=200)

plt.scatter(x=chicago_df["hh_no_internet(%)"],
            y=chicago_df["total_hispanic(%)"],
           c="red", alpha=0.5)


plt.xlabel("Percentage of Households with No Internet(%)")
plt.ylabel("Percentage of Hispanic Population in each Chicago Community Area (%)")
plt.title("Internet Access by Race (2015-19)")
```

```python
plt.figure(figsize=(15, 10), dpi=200)

plt.scatter(x=chicago_df["hh_no_internet(%)"],
            y=chicago_df["hh_unemployment(%)"],
           c="blue", alpha=0.5)


plt.xlabel("Percentage of Households with No Internet(%)")
plt.ylabel("Percentage of Households Unemployed (%)")
plt.title("Internet Access by Unemployment Rates (2015-19)")
```

```python
plt.figure(figsize=(15, 10), dpi=200)

plt.scatter(x=chicago_df["hh_no_internet(%)"],
            y=chicago_df["hh_median_income"],
           c="blue", alpha=0.5)


plt.xlabel("Percentage of Households with No Internet(%)")
plt.ylabel("Household Median Income ($)")
plt.title("Internet Access by Income (2015-19)")
```

```python
plt.figure(figsize=(15, 10), dpi=200)

plt.scatter(x=chicago_df["hh_no_internet(%)"],
            y=chicago_df["hardship_score"],
           c="purple", alpha=0.5)


plt.xlabel("Percentage of Households with No Internet(%)")
plt.ylabel("Hardship Index Score")
plt.title("Internet Access by Hardship Score (2015-19)")
```

```python
plt.figure(figsize=(15, 10), dpi=200)

plt.scatter(x=chicago_df["hh_no_internet(%)"],
            y=chicago_df["ccvi_score"],
           c="purple", alpha=0.5)


plt.xlabel("Percentage of Households with No Internet(%)")
plt.ylabel("COVID-19 Community Vulnerability Index (CCVI) Score")
plt.title("Internet Access by CCVI Score (2015-19)")
```
